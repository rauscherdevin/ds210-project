### DS210 Final - Data Cleaning
I used Python (and more specifically, the pandas package) to clean the data for my final project on Boston Marathon degrees of separation because of the number of CSV files used for the project and for my own clarity.

```python
import pandas as pd
import country_converter as coco
```

##### Winners, from https://www.kaggle.com/datasets/zhikchen/boston-marathon-winners-men-and-women

```python
winners_w = pd.read_csv('Womens_Boston_Marathon_Winners.csv')
winners_m = pd.read_csv('Mens_Boston_Marathon_Winners.csv')

print(len(winners_w))
print(len(winners_m))
```

```python
win_df = [winners_w,winners_m]
winners = pd.concat(win_df)
winners = winners.reset_index()

# deleting nan for 1918, 2022 (no marathon)
winners = winners.drop([winners.index[78],winners.index[180],winners.index[54]])

winners
```

```python
win_countries = list(winners.Country.unique())
win_countries
```

##### Finishers from 2015-2017, from https://www.kaggle.com/datasets/rojour/boston-results?select=marathon_results_2017.csv

```python
mar_2015 = pd.read_csv('marathon_results_2015.csv')
mar_2016 = pd.read_csv('marathon_results_2016.csv')
mar_2017 = pd.read_csv('marathon_results_2017.csv')
```

```python
mar_2015['Year'] = 2015
mar_2016['Year'] = 2016
mar_2017['Year'] = 2017
```

```python
other_df = [mar_2015, mar_2016, mar_2017]
other = pd.concat(other_df)
other
```

```python
other.Country.unique()
```

```python
# create a list of the countries as given
false_iso3_list = []
for country in other.Country:
    false_iso3_list.append(country)
```

```python
# unfortunately the boston marathon registration site (where the data was sourced originally) simply does not use io3 designation
# the following code fixes that to properly use the country_converter package
# checked against the marathon cite and https://en.wikipedia.org/wiki/ISO_3166-1_alpha-3
iso3_list = []
for country in false_iso3_list:
    if country=='RSA':
        new_c = 'RUS'
    elif country=='CRO':
        new_c = 'HRV'
    elif country=='GER':
        new_c = 'DEU'
    elif country=='DEN':
        new_c = 'DNK'
    elif country=='MAS':
        new_c = 'MYS'
    elif country=='CHI':
        new_c = 'CHN'
    elif country=='GUA':
        new_c = 'GTM'
    elif country=='SUI':
        new_c = 'CHE'
    elif country=='NED':
        new_c = 'NLD'
    elif country=='CRC':
        new_c = 'CRI'
    elif country=='POR':
        new_c = 'PRT'
    elif country=='TPE': # TWN in 2016, 2017; only different in 2015
        new_c = 'TWN'
    elif country=='SIN':
        new_c = 'SGP'
    elif country=='BER':
        new_c = 'BMU'
    elif country=='VIE':
        new_c = 'VNM'
    elif country=='ESA':
        new_c = 'SLV'
    elif country=='GRE':
        new_c = 'GRL'
    elif country=='SLO':
        new_c = 'SVN'
    elif country=='URU':
        new_c = 'URY'
    elif country=='CAY':
        new_c = 'CYM'
    elif country=='UAE':
        new_c = 'ARE'
    elif country=='TRI':
        new_c = 'TTO'
    elif country=='INA':
        new_c = 'IDN'
    elif country=='AHO':
        new_c = 'CUW'
    elif country=='PHI':
        new_c = 'PHL'
    elif country=='ZIM':
        new_c = 'ZWE'
    elif country=='HON':
        new_c = 'HND'
    elif country=='LAT':
        new_c = 'LVA'
    elif country=='OMA':
        new_c = 'OMN'
    elif country=='BAH':
        new_c = 'BHS'
    elif country=='KSA':
        new_c = 'SAU'
    elif country=='BUL':
        new_c = 'BGR'
    elif country=='ALG':
        new_c = 'DZA'
    elif country=='NCA':
        new_c = 'NIC'
    elif country=='BAR':
        new_c = 'BRB'
    elif country=='NGR':
        new_c = 'NGA'
    elif country=='GRN':
        new_c = 'GRL'
    elif country=='MGL':
        new_c = 'MNG'
    elif country=='PAR':
        new_c = 'PRY'
    elif country=='KUW':
        new_c = 'KWT'
    else:
        new_c = country
    iso3_list.append(new_c)
```

```python
iso3_list
```

```python
countries = coco.convert(names=iso3_list, to='name_official')
countries
```

```python
other['Country'] = countries
other
```

##### Beginning the Combination Process

```python
win_countries
## unique countries that have had winners
```

```python
## all countries from 2015-2017
tot_countries = list(other.Country.unique())
tot_countries
```

```python
## going back to other['Country'] to match names to the winning countries 
# editing countries list
renamed_countries = []
for country in countries:
    if country=='United States of America':
        temp_c = 'United States'
    elif country=='Federal Republic of Germany':
        temp_c = 'Germany'
    elif country=='Kingdom of Norway':
        temp_c = 'Norway'
    elif country=='Portuguese Republic':
        temp_c = 'Portugal'
    elif country=='Republic of Poland':
        temp_c = 'Poland'
    elif country=='Russian Federation':
        temp_c = 'Russia'
    elif country=='Federal Democratic Republic of Ethiopia':
        temp_c = 'Ethiopia'
    elif country=='Republic of Kenya':
        temp_c = 'Kenya'
    elif country=='Republic of Korea':
        temp_c = 'South Korea'
    elif country=='Kingdom of Sweden':
        temp_c = 'Sweden'
    elif country=='Republic of Guatemala':
        temp_c = 'Guatemala'
    elif country=='Republic of Finland':
        temp_c = 'Finland'
    elif country=='Kingdom of Belgium':
        temp_c = 'Belgium'
    elif country=='United Kingdom of Great Britain and Northern Ireland':
        temp_c = 'United Kingdom'
    elif country=='Republic of Colombia':
        temp_c = 'Colombia'
    elif country=='Commonwealth of Australia':
        temp_c = 'Australia'
    elif country=='Italian Republic':
        temp_c = 'Italy'
    else:
        temp_c = country
    renamed_countries.append(temp_c)
    
other['Country'] = renamed_countries
other
```

```python
## i also do not want to use all 80k rows, so for a random sample:
sample_other = other.sample(n=1000)
sample_other
```

```python
## putting everything together
df = pd.concat([winners,sample_other])
df

## getting only rows that are necessary
del df['index']
del df['Time']
del df['Distance (Miles)']
del df['Distance (KM)']
del df['Unnamed: 0']
del df['30K']
del df['35K']
del df['40K']
del df['Pace']
del df['Proj Time']
del df['Official Time']
del df['5K']
del df['10K']
del df['15K']
del df['20K']
del df['Half']
del df['25K']
del df['Overall']
del df['Division']
del df['Bib']
del df['City']
del df['State']
del df['Citizen']
del df['Unnamed: 9']
del df['Unnamed: 8']
del df['Age']
del df['Gender']
del df['M/F']
```

```python
## merge Winner and Name

# under Winner
win = df[:180]
names = []
for w in win.Winner:
    names.append(w)
names

# under Name
nam = df[180:]
names2 = []
for n in nam.Name:
    names2.append(n)
names2

# combine
names_list = names + names2
len(names_list)
```

```python
del df['Winner']

df['Name'] = names_list
df
```

```python
ones = []
for n in range (0,180):
    n = 1
    ones.append(n)
    
zeroes = []
for n in range (0,1000):
    n = 0
    zeroes.append(n)

## adding winners back
winners_bool = ones+zeroes
df['Win?'] = winners_bool
df
```

```python
year_str = []
for y in df['Year']:
    year_str.append(str(y))
year_str

df['Year'] = year_str
df
```
